## Listen und Tupel in Python

In Python sind Listen und Tupel grundlegende Datenstrukturen, die verwendet werden, um Sammlungen von Elementen zu speichern.
Sie ermöglichen es, mehrere Werte in einer einzigen Variable zu organisieren.
Obwohl sie ähnlich erscheinen, gibt es wesentliche Unterschiede in ihrer Verwendung und Eigenschaften.


### Listen

Listen sind *veränderbare* (mutable) Sequenzen. Das bedeutet, dass man nach der Erstellung einer Liste Elemente hinzufügen, entfernen oder ändern kann.
Listen werden mit eckigen Klammern `[]` definiert.


```python
# Example of creating and printing a list
my_list = [1, 2, 3, 4, 5]
print(my_list)
```

```python
# Adding an element to a list
my_list.append(6)
print(my_list)
```

```python
# Modifying an element in a list
my_list[2] = 7
print(my_list)
```

```python
# Removing an element from a list
my_list.pop(3)
print(my_list)
```

### Tupel

Tupel sind *unveränderbare* (immutable) Sequenzen. Nach der Erstellung eines Tupels können dessen Elemente nicht mehr geändert werden.
Tupel werden mit runden Klammern `()` definiert.
Sie werden oft für Datensätze verwendet, die nicht verändert werden sollen.


```python
# Example of creating and printing a tuple
my_tuple = (10, 20, 30, 40, 50)
print(my_tuple)
```

```python
# Trying to modify a tuple
my_tuple[0] = 100
print(my_tuple)
```

Weiterhin erzeugt der Befehl __range__ ein Tupel von Zahlen:

```python
t = range(10)
print(t, type(t))
```

# Packing und Unpacking

Ähnlich einfach wie das Einpacken kann man Tupel und Listen auch auspacken.

```python
datum = 26, 7, 1987
print(datum, type(datum))

tag, monat, jahr = datum
print(tag, monat, jahr)

datum = [tag, monat, jahr]
print(datum, type(datum))

tag, monat, jahr = datum
print(tag, monat, jahr)
```

### Unterschiede zwischen Listen und Tupeln

1.  **Veränderbarkeit (Mutability):** Listen sind veränderbar, Tupel sind unveränderbar.
2.  **Syntax:** Listen werden mit `[]` definiert, Tupel mit `()`.
3.  **Verwendung:** Listen werden verwendet, wenn man Sammlungen von Elementen hat, die sich ändern können.
   Tupel werden verwendet, um Datensätze zu repräsentieren, die nach der Erstellung konstant bleiben sollen (z.B. Koordinaten, Datensätze in Datenbanken).
4. **Performance:** Tupel sind im Allgemeinen etwas schneller als Listen, da sie unveränderlich sind.
