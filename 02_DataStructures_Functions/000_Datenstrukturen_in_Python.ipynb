## Datenstrukturen in Python

In Python gibt es verschiedene Datenstrukturen, die es ermöglichen, Sammlungen von Daten zu speichern und zu verwalten.
Diese Strukturen sind entscheidend für die Organisation und Verarbeitung von Informationen.
Wir werden uns hier auf Sequenztypen konzentrieren und deren Gemeinsamkeiten herausarbeiten.


### Sequenzen

Sequenzen sind geordnete Sammlungen von Elementen. 
Die Elemente in einer Sequenz haben eine bestimmte Reihenfolge, auf die durch Indizes zugegriffen werden kann.
Python bietet verschiedene Sequenztypen wie Listen, Tupel und Strings an.


#### Gemeinsame Aspekte von Sequenzen

Alle Sequenztypen teilen einige grundlegende Operationen:

- **Indexing:** Zugriff auf einzelne Elemente über ihre Position (Index). Der Index beginnt bei 0.
- **Slicing:** Extrahieren von Teilsequenzen.
- **Iteration:** Durchlaufen der Elemente in einer Sequenz mit Schleifen.
- **Länge:** Bestimmen der Anzahl der Elemente mit der Funktion `len()`.
- **Konkatenation:** Verbinden von Sequenzen mit dem `+` Operator.
- **Wiederholung:**  Erzeugen neuer Sequenzen durch Wiederholung mit dem `*` Operator.
- **Mitgliedschaft:** Prüfen, ob ein Element in der Sequenz enthalten ist mit dem `in` Operator.


```python
# Examples for different sequence types

# Lists are mutable sequences
my_list = [10, 20, 30, 40, 50]

# Tuples are immutable sequences
my_tuple = (1, 2, 3, 4, 5)

# Strings are also immutable sequences
my_string = "Hallo Welt"
```

### Indexing

Hier zeigen wir Beispiele für Indexing, um auf einzelne Elemente zuzugreifen.


```python
# Accessing elements by index
print(f"First element of list: {my_list[0]}")
print(f"First element of tuple: {my_tuple[0]}")
print(f"First character of string: {my_string[0]}")

# Accessing last elements by negative indexing
print(f"Last element of list: {my_list[-1]}")
print(f"Last element of tuple: {my_tuple[-1]}")
print(f"Last character of string: {my_string[-1]}")
```

### Slicing

Slicing ermöglicht es, einen Teil einer Sequenz zu extrahieren.


__Die Grund-Syntax__

Das Slicing wird mit eckigen Klammern [] durchgeführt und besteht aus bis zu drei Werten, die durch Doppelpunkte : getrennt sind:

    sequenz[start:stop:step]

- start: Der Index, bei dem der Slice beginnt (__inklusive__). Wenn du ihn weglässt, beginnt Python standardmäßig bei 0 (dem Anfang der Sequenz).

- stop: Der Index, bei dem der Slice endet (__exklusive!__ Das Element an dieser Stelle ist nicht mehr enthalten). Wenn du ihn weglässt, geht Python standardmäßig bis zum Ende der Sequenz.

- step: Die Schrittweite (optional). Bestimmt, wie viele Elemente übersprungen werden. Der Standardwert ist 1 (jedes Element).


```python
# Slicing sequences
print(f"Slice of list: {my_list[1:4]}")
print(f"Slice of tuple: {my_tuple[1:4:2]}")
print(f"Slice of string: {my_string[1:7]}")
```

```python
# Negative indexing
print(my_string[-4:])    
print(my_list[:-3])

# Negative stepping
print(my_string[::-1])    

```

### Iteration

Wir können mit einer `for`-Schleife durch Sequenzen iterieren.


```python
# Iteration over sequences
print("List elements:")
for element in my_list:
    print(element)

print("Tuple elements:")
for element in my_tuple:
    print(element)

print("String characters:")
for char in my_string:
    print(char)
```

### Länge

Die Funktion `len()` gibt die Anzahl der Elemente in einer Sequenz zurück.


```python
# Using len() to get the length
print(f"Length of list: {len(my_list)}")
print(f"Length of tuple: {len(my_tuple)}")
print(f"Length of string: {len(my_string)}")
```

### Konkatenation

Sequenzen können mit dem `+` Operator verkettet werden.


```python
# Concatenation of sequences
new_list = my_list + [60, 70]
new_tuple = my_tuple + (6, 7)
new_string = my_string + "! wie geht es?"

print(new_list)
print(new_tuple)
print(new_string)
```

### Wiederholung

Sequenzen können mit dem `*` Operator wiederholt werden.


```python
# Repetition of sequences
repeated_list = my_list * 2
repeated_tuple = my_tuple * 2
repeated_string = my_string * 2

print(repeated_list)
print(repeated_tuple)
print(repeated_string)
```

### Mitgliedschaft

Überprüfen ob ein Element in einer Sequenz vorhanden ist, mit dem `in` Operator.


```python
# Membership testing
print(f"20 in list: {20 in my_list}")
print("20 in list: ", 20 in my_list)
print(f"6 in tuple: {6 in my_tuple}")
print(f"e in string: {'e' in my_string}")
```

### Fazit

Sequenzen sind grundlegende Datenstrukturen in Python mit gemeinsamen Operationen. 
Das Verständnis dieser Gemeinsamkeiten ermöglicht es uns, effizienter mit Listen, Tupeln und Strings zu arbeiten.
