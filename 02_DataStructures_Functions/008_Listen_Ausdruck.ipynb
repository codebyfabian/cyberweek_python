# Listen-Ausdrücke in Python
Listen-Ausdrücke, auch als "list comprehensions" bekannt, sind eine prägnante Methode, um neue Listen zu erstellen, indem man eine bestehende Liste oder Sequenz iteriert und optional Filterbedingungen anwendet. Sie folgen der mathematischen Notation für Mengen, sind jedoch in Python viel effizienter, sowohl in der Ausführung als auch im Schreibaufwand.

## Grundlegende Syntax
Der grundlegende Aufbau eines Listen-Ausdrucks sieht wie folgt aus:

```python
[expression for item in iterable]

[expression for item in iterable if condition]
```

- `expression` ist der Ausdruck, der auf jedes Element angewendet wird.
- `item` ist ein Platzhalter für jedes Element in der `iterable`.
- `iterable` ist die Liste oder Sequenz, die durchlaufen wird.
- `condition` ist eine optionale Bedingung, die ein Element erfüllen muss, um in die neue Liste aufgenommen zu werden.

## Einfache Beispiele

````python
# Erstelle eine Liste der Quadrate von 0 bis 9
quadrate = [x**2 for x in range(10)]
print(quadrate)
````

````python
# Erstellt eine Liste mit den Zeichen in einem String
zeichen = [buchstabe for buchstabe in 'Hallo']
print(zeichen)
````

## Filtern mit Listen-Ausdrücken
Beim Filtern verwenden wir die `if`-Klausel, um nur die Elemente auszuwählen, die eine Bedingung erfüllen.

````python
# Nur gerade Zahlen von 0 bis 9 in die Liste aufnehmen
gerade = [x for x in range(10) if x % 2 == 0]
print(gerade)
````

## Mapping mit Listen-Ausdrücken
Mapping bedeutet, dass wir auf jedes Element einer Liste eine Funktion oder einen Ausdruck anwenden, um eine neue Liste mit den Ergebnissen zu erstellen.

````python
# Verdopple jeden Wert in der Liste
zahlen = [1, 2, 3, 4, 5]
verdoppelte_zahlen = [x * 2 for x in zahlen]
print(verdoppelte_zahlen)
````

## Kombination von Filtern und Mapping
Wir können Filtern und Mapping kombinieren, um mächtigere Listen-Ausdrücke zu erstellen.

````python
# Quadrat von geraden Zahlen von 0 bis 9
quadrate_gerade = [x**2 for x in range(10) if x % 2 == 0]
print(quadrate_gerade)
````
