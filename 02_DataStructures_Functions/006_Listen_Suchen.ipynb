## Listen in Python
Listen sind eine der grundlegendsten Datenstrukturen in Python. Sie sind geordnete Sammlungen von Elementen, die unterschiedliche Datentypen enthalten können.
Hier sind einige wichtige Punkte zu Listen:
*   **Erstellung:** Listen werden mit eckigen Klammern `[]` erstellt.
*   **Veränderbarkeit:** Listen sind veränderbar, d.h., man kann Elemente hinzufügen, entfernen oder ändern.
*   **Indizierung:** Elemente in einer Liste sind über ihren Index zugänglich, wobei der erste Index 0 ist.
*   **Duplikate:** Listen können doppelte Elemente enthalten.


```python
# Beispiel einer Liste mit verschiedenen Datentypen
my_list = [1, "hello", 3.14, True]
print(my_list)
```

### Zugriff auf Listenelemente
Elemente in einer Liste werden über ihren Index abgerufen. Der Index beginnt bei 0 für das erste Element.


```python
# Zugriff auf Elemente mit Index
print(my_list[0])  # Gibt 1 aus
print(my_list[1])  # Gibt 'hello' aus
print(my_list[-1]) # Gibt True aus. Negative Indizes zählen vom Ende der Liste
```

### Suchen in Listen
Python bietet verschiedene Möglichkeiten, Elemente in einer Liste zu suchen.


```python
# Suchen mit der `in` Methode
print('hello' in my_list)  # Gibt True aus, da 'hello' in der Liste ist
print(5 in my_list)      # Gibt False aus, da 5 nicht in der Liste ist
#  Mit der index Methode
try:
    index_of_hello = my_list.index('hello')
    print(f'Index von "hello": {index_of_hello}')
    index_of_5 = my_list.index(5) # Dies führt zu einem ValueError
except ValueError:
    print('5 ist nicht in der Liste')
```

### Ersetzen von Elementen in Listen
Elemente in einer Liste können direkt über ihren Index ersetzt werden.


```python
# Element ersetzen
my_list[1] = 'world'
print(my_list)  # Die Liste ist nun [1, 'world', 3.14, True]
```

### Löschen von Elementen in Listen
Es gibt mehrere Möglichkeiten, Elemente aus einer Liste zu löschen.


```python
# Löschen mit `del` und Index
del my_list[0]
print(my_list)  # Die Liste ist nun ['world', 3.14, True]
# Löschen mit `remove` und Wert
my_list.remove(3.14)
print(my_list) # Die Liste ist nun ['world', True]
# Löschen mit `pop` und Index
popped_element = my_list.pop(0)
print(f'Element {popped_element} wurde entfernt')
print(my_list) # Die Liste ist nun [True]
#  Leeren der Liste
my_list.clear()
print(my_list) # Die Liste ist nun []
```

### Weitere Listenoperationen
Es gibt viele weitere Operationen, die mit Listen durchgeführt werden können, wie z.B. das Hinzufügen von Elementen mit `append`, `extend` oder `insert`, sowie das Sortieren von Listen mit `sort`.
