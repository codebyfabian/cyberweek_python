**Aufgabe 1: Flexibler Addierer**

Schreibe eine Funktion `flexible_adder`, die eine beliebige Anzahl von numerischen Argumenten akzeptiert und ihre Summe zurückgibt.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def flexible_adder(*args):
    """Adds any number of numeric arguments.
    
    Args:
        *args: Arbitrary number of numeric arguments.
    Returns:
        Sum of the arguments.
    """
    total = 0
    for num in args:
        total += num
    return total

print(flexible_adder(1, 2, 3))
print(flexible_adder(5, 10, 15, 20))
print(flexible_adder())
```

**Aufgabe 2:  String-Konkatenator**
Erstelle eine Funktion `string_concat`, die eine beliebige Anzahl von Strings als Argumente akzeptiert und einen einzigen String zurückgibt, der aus der Konkatenation aller Eingabestrings besteht. 


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def string_concat(*args):
    """Concatenates any number of string arguments.
    
    Args:
        *args: Arbitrary number of string arguments.
    Returns:
        Concatenation of the strings.
    """
    result = ""
    for s in args:
        result += s
    return result

print(string_concat('Hello', ', ', 'World', '!'))
print(string_concat('Python', ' is', ' awesome'))
print(string_concat())
```

**Aufgabe 3:  Anzahl der Elemente**
Schreibe eine Funktion `count_elements`, die eine beliebige Anzahl von Argumenten annimmt und ein Dictionary zurückgibt, in dem die Anzahl der Vorkommen jedes Arguments gezählt wird.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def count_elements(*args):
    """Counts the occurrences of each argument.
    
    Args:
        *args: Arbitrary number of arguments.
    Returns:
        Dictionary of counts of each argument.
    """
    counts = {}
    for item in args:
        if item in counts:
            counts[item] += 1
        else:
            counts[item] = 1
    return counts

print(count_elements(1, 2, 2, 3, 3, 3))
print(count_elements('a', 'b', 'a', 'c', 'b'))
print(count_elements())
```

**Aufgabe 4:  Multiplikator-Closure**
Schreibe eine Funktion `multiplier_generator`, die einen Faktor als Argument nimmt und eine Closure zurückgibt. Die Closure soll eine Funktion sein, die eine Zahl als Argument nimmt und das Produkt aus der Zahl und dem Faktor zurückgibt.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def multiplier_generator(factor):
    """Generates a closure that multiplies a number by a factor.
    
    Args:
        factor: The multiplication factor.
    Returns:
        A closure function.
    """
    def multiplier(number):
        return number * factor
    return multiplier

double = multiplier_generator(2)
triple = multiplier_generator(3)

print(double(5))
print(triple(5))
print(double(10))
```

**Aufgabe 5:  Potenzierer-Closure**
Erstelle eine Funktion `power_generator`, die einen Exponenten als Argument nimmt und eine Closure zurückgibt. Die Closure soll eine Funktion sein, die eine Basis als Argument nimmt und die Basis mit dem Exponenten potenziert.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def power_generator(exponent):
    """Generates a closure that raises a base to the given exponent.
    
    Args:
        exponent: The exponent.
    Returns:
        A closure function.
    """
    def power(base):
        return base ** exponent
    return power

square = power_generator(2)
cube = power_generator(3)

print(square(4))
print(cube(2))
print(square(5))
```

**Aufgabe 6:  Funktions-Decorator für Zeitmessung**
Schreibe einen Decorator `time_it`, der die Ausführungszeit einer Funktion misst und ausgibt. Die zu dekorierende Funktion kann beliebige Parameter haben.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
import time

def time_it(func):
    """Decorator to measure the execution time of a function.
    
    Args:
        func: The function to decorate.
    Returns:
        The decorated function.
    """
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@time_it
def slow_function(n):
    time.sleep(n)
    return 'done'

slow_function(2)
slow_function(1)
```

**Aufgabe 7:  Funktions-Decorator für Cache**
Schreibe einen Decorator `cache_it`, der die Ergebnisse einer Funktion speichert (caching). Die zu dekorierende Funktion kann beliebige Parameter haben. Verwende ein Dictionary zum Speichern der Ergebnisse.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def cache_it(func):
    """Decorator to cache results of a function.
    
    Args:
        func: The function to decorate.
    Returns:
        The decorated function.
    """
    cache = {}
    def wrapper(*args):
        if args in cache:
            print('From Cache!')
            return cache[args]
        else:
            print('Not from Cache')
            result = func(*args)
            cache[args] = result
            return result
    return wrapper

@cache_it
def expensive_function(x, y):
    print('calculating...')
    return x + y

print(expensive_function(2, 3))
print(expensive_function(2, 3))
print(expensive_function(3, 4))
```

**Aufgabe 8:  Logger-Decorator**
Schreibe einen Decorator `log_it`, der die Argumente und den Rückgabewert einer Funktion in der Konsole protokolliert. Die zu dekorierende Funktion kann beliebige Parameter haben.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def log_it(func):
    """Decorator to log arguments and return value of a function.
    
    Args:
        func: The function to decorate.
    Returns:
        The decorated function.
    """
    def wrapper(*args, **kwargs):
        print(f'Calling function {func.__name__} with args: {args} and kwargs: {kwargs}')
        result = func(*args, **kwargs)
        print(f'Function {func.__name__} returned {result}')
        return result
    return wrapper

@log_it
def sample_function(x, y=10):
    return x * y

sample_function(5)
sample_function(5,y=2)
```

**Aufgabe 9:  Zähler-Closure**
Implementiere eine Funktion `counter_generator`, die eine Closure zurückgibt. Die Closure soll einen Zähler inkrementieren und den aktuellen Wert zurückgeben.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def counter_generator():
    """Generates a closure that increments a counter.
    
    Returns:
        A closure function.
    """
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

counter1 = counter_generator()
counter2 = counter_generator()

print(counter1())
print(counter1())
print(counter2())
print(counter1())
```

**Aufgabe 10:  Konfigurierbare Logger-Closure**
Schreibe eine Funktion `configurable_logger`, die eine Log-Nachricht und einen Präfix entgegennimmt und eine Closure zurückgibt. Die Closure soll eine Funktion sein, die eine Nachricht entgegennimmt, die beiden Nachrichten zusammenfügt und mit dem Präfix versieht und dann ausgibt.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
def configurable_logger(prefix, log_message):
    """Generates a configurable logger closure.
    
    Args:
        prefix: The prefix for the log message.
        log_message: The base log message.
    Returns:
        A closure function.
    """
    def logger(message):
        print(f'{prefix} : {log_message} {message}')
    return logger

log_info = configurable_logger('INFO', 'Process completed:')
log_error = configurable_logger('ERROR', 'An error occurred:')

log_info('Task 1')
log_error('File not found')
log_info('Task 2')
```
