## Arbeiten mit mehreren Modulen in Python

In diesem Notebook werden wir untersuchen, wie man mit mehreren Modulen in Python arbeitet. Wir werden uns darauf konzentrieren, wie man Funktionen und Variablen zwischen Modulen exportiert und importiert, und dies anhand von Beispielen veranschaulichen.


### Modul erstellen (my_module.py)
Zuerst erstellen wir ein einfaches Modul namens `my_module.py`, das einige Funktionen und Variablen enthält.


```python
# my_module.py

# Define a variable
my_variable = 10

# Define a function
def my_function(x):
    return x * 2

# Define another function
def another_function(y):
    return y + 5

# Define a class
class MyClass:
    def __init__(self, value):
        self.value = value

    def get_value(self):
        return self.value
```

### Importieren des Moduls
Es gibt verschiedene Möglichkeiten, ein Modul zu importieren und auf seine Inhalte zuzugreifen.


#### 1. `import my_module`
Importiert das gesamte Modul.  Auf die Elemente des Moduls kann mit `my_module.element` zugegriffen werden.


```python
import my_module

# Access the variable
print(my_module.my_variable)

# Call a function
print(my_module.my_function(5))

# Create an instance of the class
my_instance = my_module.MyClass(20)
print(my_instance.get_value())
```

#### 2. `from my_module import my_function, my_variable`
Importiert spezifische Elemente aus dem Modul. Auf diese Elemente kann direkt zugegriffen werden, ohne den Modulnamen voranzustellen.


```python
from my_module import my_function, my_variable, MyClass

# Access the variable directly
print(my_variable)

# Call the function directly
print(my_function(5))

# Create an instance of the class directly
my_instance = MyClass(30)
print(my_instance.get_value())
```

#### 3. `from my_module import *`
Importiert alle Elemente aus dem Modul.  Dies ist im Allgemeinen nicht empfohlen, da es zu Namenskonflikten führen kann.


```python
from my_module import *

# Access the variable directly
print(my_variable)

# Call the function directly
print(my_function(5))

print(another_function(10))

# Create an instance of the class directly
my_instance = MyClass(40)
print(my_instance.get_value())
```

#### 4. `import my_module as mm`
Importiert das Modul und vergibt einen Alias. Dies ist nützlich, um Namenskonflikte zu vermeiden oder um den Code lesbarer zu machen.


```python
import my_module as mm

# Access the variable using the alias
print(mm.my_variable)

# Call a function using the alias
print(mm.my_function(5))

# Create an instance of the class using the alias
my_instance = mm.MyClass(50)
print(my_instance.get_value())
```

#### 5. `from my_module import my_function as mf`
Importiert ein spezifisches Element und vergibt einen Alias.


```python
from my_module import my_function as mf

# Call the function using the alias
print(mf(5))
```

### Zusammenfassung

Dieses Notebook hat verschiedene Möglichkeiten gezeigt, wie man Module in Python importiert und verwendet.  Die Wahl der Methode hängt von den spezifischen Anforderungen des Projekts und den Vorlieben des Entwicklers ab.
