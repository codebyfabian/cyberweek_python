## Listen-Funktionen in Python

In diesem Notebook werden wir uns einige nützliche Funktionen im Zusammenhang mit Listen ansehen, insbesondere `map()` und `enumerate()`.


### `map()`-Funktion

Die `map()`-Funktion wendet eine Funktion auf jedes Element einer iterierbaren Struktur (z. B. eine Liste) an und gibt ein neues iterierbares Objekt zurück.
Die allgemeine Syntax ist: `map(function, iterable, ...)`



```python
# Beispiel 1: Quadrieren aller Elemente einer Liste

numbers = [1, 2, 3, 4, 5]

def square(x):
    return x ** 2

squared_numbers = map(square, numbers)

# Beachte: map() gibt ein map-Objekt zurück, das wir in eine Liste umwandeln können.
squared_numbers_list = list(squared_numbers)

print(squared_numbers_list) # Ausgabe: [1, 4, 9, 16, 25]
```

Wir können auch eine `lambda`-Funktion verwenden, um den Code kompakter zu gestalten:


```python
# Beispiel 2: Quadrieren mit lambda-Funktion

numbers = [1, 2, 3, 4, 5]

squared_numbers = map(lambda x: x ** 2, numbers)
squared_numbers_list = list(squared_numbers)

print(squared_numbers_list) # Ausgabe: [1, 4, 9, 16, 25]
```

```python
# Beispiel 3: Verwendung von map() mit mehreren Iterables

list1 = [1, 2, 3]
list2 = [4, 5, 6]

sum_lists = map(lambda x, y: x + y, list1, list2)
sum_lists_list = list(sum_lists)

print(sum_lists_list)  # Ausgabe: [5, 7, 9]
```

### `enumerate()`-Funktion

Die `enumerate()`-Funktion fügt einen Zähler zu einem Iterable hinzu und gibt ein Enumerate-Objekt zurück.  Dieses Objekt enthält Tupel der Form (Index, Element).
Die allgemeine Syntax ist: `enumerate(iterable, start=0)`



```python
# Beispiel 1: Iterieren über eine Liste mit Index

fruits = ['apple', 'banana', 'cherry']

enum = [(0,'apple'),(1,'banana'),(2,'cherry')]

for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")
# Ausgabe:
# Index: 0, Fruit: apple
# Index: 1, Fruit: banana
# Index: 2, Fruit: cherry
```

```python
# Beispiel 2: Startindex ändern

fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits, start=1):
    print(f"Index: {index}, Fruit: {fruit}")
# Ausgabe:
# Index: 1, Fruit: apple
# Index: 2, Fruit: banana
# Index: 3, Fruit: cherry
```

```python
# Beispiel 3: Enumerate in list comprehension

fruits = ['apple', 'banana', 'cherry']

indexed_fruits = [(index, fruit) for index, fruit in enumerate(fruits)]

print(indexed_fruits)
# Ausgabe: [(0, 'apple'), (1, 'banana'), (2, 'cherry')]#
```
