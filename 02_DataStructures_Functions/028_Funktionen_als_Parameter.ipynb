## Funktionen als Parameter
In Python können Funktionen als Parameter an andere Funktionen übergeben werden. Das ermöglicht sehr flexible und wiederverwendbare Code-Strukturen.

Wir werden uns folgende Konzepte genauer ansehen:
- Callbacks
- Funktionen in Funktionen
- Funktionen als Rückgabewert
- Closures


### Callbacks
Eine Callback-Funktion ist eine Funktion, die als Argument an eine andere Funktion übergeben wird. Die aufrufende Funktion ruft dann die Callback-Funktion zu einem bestimmten Zeitpunkt auf. Callbacks werden oft in asynchronen Operationen oder bei der Ereignisverarbeitung verwendet.


```python
# Example of a callback function

def apply_operation(x, y, operation):
    # Apply the operation to x and y
    return operation(x, y)

def add(x, y):
    # Add two values
    return x + y

def subtract(x, y):
    # Subtract two values
    return x - y

# Using callback functions
result_add = apply_operation(5, 3, add)
result_subtract = apply_operation(5, 3, subtract)

print(f"Addition result: {result_add}") # Output: Addition result: 8
print(f"Subtraction result: {result_subtract}") # Output: Subtraction result: 2
```

### Funktionen in Funktionen
In Python können Funktionen innerhalb anderer Funktionen definiert werden. Die innere Funktion ist dann nur im Scope der äußeren Funktion sichtbar (lokaler Scope). Dies ist nützlich, um Hilfsfunktionen zu kapseln.


```python
# Example of a function inside another function

def outer_function(x):
    def inner_function(y):
        # Inner function multiplies two numbers
        return x * y
    # Return inner function's result
    return inner_function

# create inner_function with predefined x=5
multiplier = outer_function(5)
result_inner = multiplier(3)
print(f"Result of the inner function: {result_inner}") # Output: Result of the inner function: 15
```

### Funktionen als Rückgabewert
Funktionen können in Python auch als Rückgabewert einer anderen Funktion zurückgegeben werden. Das ermöglicht die Erzeugung dynamischer Funktionen.


```python
# Example of a function as a return value

def create_operation(operation_type):
    if operation_type == "add":
        def add(x, y):
            # Returns the sum of x and y
            return x + y
        return add
    elif operation_type == "subtract":
        def subtract(x, y):
            # Returns the difference of x and y
            return x - y
        return subtract
    else:
       return None

# Function returned as result
add_function = create_operation("add")
subtract_function = create_operation("subtract")

if add_function:
    result_add = add_function(5, 3)
    print(f"Addition result: {result_add}") # Output: Addition result: 8
if subtract_function:
    result_subtract = subtract_function(5, 3)
    print(f"Subtraction result: {result_subtract}") # Output: Subtraction result: 2
```

### Closures
Eine Closure ist eine innere Funktion, die auf Variablen aus dem Scope der äußeren Funktion zugreift, auch nachdem die äußere Funktion beendet wurde.  Die innere Funktion 'erinnert' sich an die Umgebung, in der sie erzeugt wurde.


```python
# Example of a closure

def create_multiplier(factor):
    def multiplier(x):
        # Inner function multiplies by the captured factor
        return x * factor
    return multiplier

# create closures
multiply_by_2 = create_multiplier(2)
multiply_by_3 = create_multiplier(3)

print(f"Result of multiply_by_2 with 5: {multiply_by_2(5)}") # Output: Result of multiply_by_2 with 5: 10
print(f"Result of multiply_by_3 with 5: {multiply_by_3(5)}") # Output: Result of multiply_by_3 with 5: 15
```
