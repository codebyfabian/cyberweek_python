### Zeichenketten (Strings)
Eine Zeichenkette, oder String, ist eine Sequenz von Zeichen. Strings sind in Python unveränderlich (immutable), d.h. man kann sie nach ihrer Erstellung nicht mehr direkt verändern.
Strings werden verwendet, um Text darzustellen.
Strings können mit einfachen oder doppelten Anführungszeichen erstellt werden.


```python
# String creation using single quotes
single_quoted_string = 'This is a string using single quotes'
# String creation using double quotes
double_quoted_string = "This is a string using double quotes"
print(single_quoted_string)
print(double_quoted_string)
```

Strings können auch mehrzeilige Texte enthalten. Hierfür werden dreifache Anführungszeichen verwendet, entweder einfach oder doppelt.


```python
# Multiline string using triple single quotes
multiline_string_single = '''This is a multiline
string using triple single quotes.'''
# Multiline string using triple double quotes
multiline_string_double = """This is a multiline
string using triple double quotes."""
print(multiline_string_single)
print(multiline_string_double)
```

### String-Operationen
Python bietet eine Vielzahl von Operationen, die auf Strings angewendet werden können, wie z.B. String-Konkatenation, String-Slicing und String-Formatierung.
Hier sind einige Beispiele:

```python
# String concatenation
str1 = 'Hello'
str2 = 'World'
concatenated_string = str1 + ' ' + str2
print(concatenated_string) # Output: Hello World

# String slicing
my_string = 'Python is awesome'
sliced_string = my_string[0:6] # get substring from index 0 up to (but not including) index 6
print(sliced_string) # Output: Python

# String length
print(len(my_string)) # Output: 17

# String formatting
name = 'Alice'
age = 30
formatted_string = f'My name is {name} and I am {age} years old.'
print(formatted_string) # Output: My name is Alice and I am 30 years old.
```
