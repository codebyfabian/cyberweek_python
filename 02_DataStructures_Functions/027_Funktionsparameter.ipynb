## Funktionsparameter in Python
In Python sind Funktionsparameter die Variablen, die in der Funktionsdefinition in den Klammern nach dem Funktionsnamen stehen.
Sie dienen dazu, Werte an die Funktion zu übergeben, die dann innerhalb der Funktion verwendet werden können.
Es gibt verschiedene Arten von Parametern, die wir uns genauer ansehen werden.


```python
# Simple function with parameters
def greet(name, greeting):
    print(greeting + ", " + name + "!")

greet("Alice", "Hello")  # Output: Hello, Alice!
greet("Bob", "Hi")      # Output: Hi, Bob!
```

### Default-Parameter
Default-Parameter sind Parameter, denen ein Standardwert zugewiesen wird.
Wenn beim Aufruf der Funktion kein Wert für diesen Parameter angegeben wird,
wird der Standardwert verwendet.


```python
# Function with default parameter
def greet_default(name, greeting="Hello"):
    print(greeting + ", " + name + "!")

greet_default("Alice")        # Output: Hello, Alice!
greet_default("Bob", "Hi")  # Output: Hi, Bob!
```

### Variable Anzahl von Parametern (*args)
Manchmal ist es nützlich, eine Funktion zu haben, die eine variable Anzahl von Parametern entgegennehmen kann.
Dies wird mit dem `*args` Parameter ermöglicht. Die Parameter werden dann als Tupel innerhalb der Funktion verfügbar.


```python
# Function with variable number of parameters
def print_args(*args):
    for arg in args:
        print(arg)

print_args(1, 2, 3) # Output: 1, 2, 3 each on a new line
print_args("a", "b", "c", "d") # Output: a, b, c, d each on a new line
print_args() # Output: (nothing will be printed)
```

### Schlüsselwortparameter (**kwargs)
Schlüsselwortparameter ermöglichen es, einer Funktion eine variable Anzahl von benannten Parametern zu übergeben.
Diese werden als Dictionary in der Funktion verfügbar.


```python
# Function with keyword parameters
def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="Alice", age=30) # Output: name: Alice, age: 30
print_kwargs(city="New York", country="USA") # Output: city: New York, country: USA
```

### Kombination von Parametertypen
Es ist auch möglich, alle Arten von Parametern in einer Funktion zu kombinieren.
Dabei ist die Reihenfolge wichtig: Zuerst kommen die normalen Parameter, dann `*args` und zuletzt `**kwargs`.


```python
# Example of combining parameters
def combined_params(a, b, *args, **kwargs):
    print(f"a: {a}, b: {b}")
    print("Args:", args)
    print("Kwargs:", kwargs)

combined_params(1, 2, 3, 4, name="Alice", age=30) 
# Output: a: 1, b: 2
#         Args: (3, 4)
#         Kwargs: {'name': 'Alice', 'age': 30}
```
