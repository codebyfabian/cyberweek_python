## Gültigkeitsbereiche in Python

In Python bezieht sich der Gültigkeitsbereich auf den Bereich eines Programms, in dem ein Name (Variable, Funktion usw.) sichtbar und zugänglich ist.
Es ist entscheidend, um Namenskonflikte zu vermeiden und die Organisation von Code zu verbessern.
Python verwendet das LEGB-Regel, um die Reihenfolge festzulegen, in der Gültigkeitsbereiche durchsucht werden:
*   **L**ocal: Gültigkeitsbereich innerhalb einer Funktion
*   **E**nclosing function locals: Gültigkeitsbereich in umschließenden Funktionen (z. B. Closures)
*   **G**lobal: Gültigkeitsbereich auf Modulebene
*   **B**uilt-in: Vordefinierte Namen in Python



### Lokale Variablen
Lokale Variablen werden innerhalb einer Funktion definiert und sind nur innerhalb dieser Funktion zugänglich.
Sie existieren nur, während die Funktion ausgeführt wird.



```python
def my_function():
    local_variable = 10  # local_variable ist lokal
    print(f"Inside the function: {local_variable}")

my_function()
# print(local_variable)  # Dies würde einen Fehler verursachen, da local_variable außerhalb der Funktion nicht existiert
```

### Globale Variablen
Globale Variablen werden außerhalb von Funktionen definiert und sind im gesamten Programm zugänglich, einschließlich innerhalb von Funktionen.



```python
global_variable = 20  # global_variable ist global

def my_function():
    print(f"Inside the function: {global_variable}")

my_function()
print(f"Outside the function: {global_variable}")

# Modifizieren einer globalen Variable innerhalb einer Funktion:
def modify_global():
    global global_variable  # Deklariert, dass wir die globale Variable verwenden wollen
    global_variable = 30
    print(f"Modified inside function: {global_variable}")

modify_global()
print(f"Modified outside function: {global_variable}")
```

### Nonlocal Variablen und Closures
Eine Closure ist eine Funktion, die Zugriff auf Variablen aus ihrem umgebenden (nicht-globalen) Gültigkeitsbereich hat, auch nachdem die äußere Funktion ihre Ausführung abgeschlossen hat.
Das Schlüsselwort `nonlocal` wird verwendet, um anzuzeigen, dass eine Variable nicht lokal, sondern im Gültigkeitsbereich einer umschließenden Funktion definiert ist.



```python
def outer_function():
    outer_variable = 100

    def inner_function():
        nonlocal outer_variable  # Deklariert, dass outer_variable aus dem äußeren Gültigkeitsbereich stammt
        outer_variable = 110
        print(f"Inside inner function: {outer_variable}")

    inner_function()
    print(f"Inside outer function: {outer_variable}")

outer_function()

# Beispiel einer Closure:
def multiplier(factor):
    def inner(number):
        return number * factor  # 'factor' wird aus dem äußeren Gültigkeitsbereich übernommen
    return inner

multiply_by_5 = multiplier(5)
print(multiply_by_5(3))  # Gibt 15 aus (3 * 5)

multiply_by_10 = multiplier(10)
print(multiply_by_10(3)) # Gibt 30 aus (3 * 10)
```

### Zusammenfassung
*   **Lokale Variablen:** Beschränkt auf die Funktion, in der sie definiert sind.
*   **Globale Variablen:** Zugänglich im gesamten Programm.
*   **Nonlocal Variablen:**  Ermöglichen den Zugriff und die Modifikation von Variablen in umschließenden Funktionen (wichtig für Closures).
*   **Closures:** Funktionen, die sich an den Zustand ihrer umgebenden Gültigkeitsbereiche erinnern, selbst nachdem die äußere Funktion beendet ist.
