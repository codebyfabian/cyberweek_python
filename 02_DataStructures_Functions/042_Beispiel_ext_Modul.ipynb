# Beispiel zur Verwendung eines Moduls

Im folgenden Beispiel werden die Module __urllib__ und __json__ verwendet.

Zunächst müssen zwei Module importiert werden:


```python
import urllib.request
import json
```

Die folgende Funktion verwendet die __urllib__ um eine Resource aus dem Internet zu laden.

```python
def get_decode(urlstring):
    with urllib.request.urlopen(urlstring) as url:
        content = url.read().decode()
        unpacked = json.loads(content)
        return unpacked
```

Der Aufruf erfolgt mit einer URL (unified resource locator), die mit einer formatierten Zeichenkette zusammengestellt wird:

```python
year = 2026
table = get_decode(f"https://www.openligadb.de/api/getbltable/bl1/{year}")
```

```python
# Erlaubt Punkt-Zugriff auf Dictionaries
class DotDict(dict):
    __getattr__ = dict.get
    __setattr__ = dict.__setitem__
    __delattr__ = dict.__delitem__
```

```python
for idx, team in enumerate(table, 1):
    # print(team.keys())
    oteam = DotDict(team)
    # print(team["teamName"])
    print(f"{idx:2}: {oteam.teamName:25} {oteam.points:2}")
```
