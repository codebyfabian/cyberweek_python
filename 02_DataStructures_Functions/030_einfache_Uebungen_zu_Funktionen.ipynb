### Aufgabe 10: Funktion zur Berechnung der Fakultät
Schreibe eine Funktion `factorial`, die eine nicht-negative ganze Zahl als Parameter entgegennimmt und die Fakultät dieser Zahl berechnet und zurückgibt. Zum Beispiel sollte `factorial(5)` den Wert 120 zurückgeben.


```python
# Ihre Loesung
```

### Musterlösung
Bitte erst öffnen, wenn Sie eine eigene Lösung haben.

```python
# Musterloesung
# Loesung fuer Aufgabe 10
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Test der Funktion
print(factorial(5))
```
