# Einführung in Variablen und Operatoren in Python\n\nDieses Notebook bietet eine ausführliche Einführung in die Welt der Variablen in Python. Wir behandeln die Zuweisung von Werten, Namenskonventionen sowie die verschiedenen Arten von Operatoren.

## 1. Was sind Variablen?\n\nIn Python dienen Variablen als Container zum Speichern von Datenwerten. Im Gegensatz zu vielen anderen Programmiersprachen muss in Python der Datentyp einer Variable nicht explizit deklariert werden (dynamische Typisierung). Eine Variable wird in dem Moment erstellt, in dem ihr zum ersten Mal ein Wert zugewiesen wird.

```python
# Einfache Zuweisung\nname = \"Max Mustermann\"\nalter = 25\ngroesse = 1.85\n\nprint(\"Name:\", name)\nprint(\"Alter:\", alter)\nprint(\"Größe:\", groesse)
```

### Namenskonventionen\n- Variablennamen müssen mit einem Buchstaben oder dem Unterstrich `_` beginnen.\n- Sie dürfen nur alphanumerische Zeichen und Unterstriche enthalten (A-z, 0-9, und _).\n- Variablennamen sind **case-sensitive** (Groß- und Kleinschreibung wird unterschieden): `alter`, `Alter` und `ALTER` sind drei verschiedene Variablen.\n- Reservierte Schlüsselwörter (wie `if`, `else`, `while`, `True`) dürfen nicht als Namen verwendet werden.

## 2. Zuweisungen (Assignments)\n\nPython ermöglicht auch komplexe Zuweisungsmuster.

```python
# Mehrfache Zuweisung in einer Zeile\na, b, c = 5, 10, 15\nprint(f\"a: {a}, b: {b}, c: {c}\")\n\n# Den gleichen Wert mehreren Variablen zuweisen\nx = y = z = 100\nprint(f\"x: {x}, y: {y}, z: {z}\")\n\n# Dynamische Typisierung: Der Typ kann sich ändern\ndaten = 42\nprint(\"Daten als Zahl:\", daten, type(daten))\ndaten = \"Jetzt eine Zeichenkette\"\nprint(\"Daten als String:\", daten, type(daten))
```
